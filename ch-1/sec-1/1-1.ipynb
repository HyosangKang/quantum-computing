```python
import numpy as np 
import matplotlib.pyplot as plt

z1, z2 = complex(1, 1), complex(0, 2)
z3 = z1 + z2
quiver_params = dict(angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, z1.real, z1.imag, **quiver_params)
plt.quiver(0, 0, z2.real, z2.imag, **quiver_params)
plt.quiver(0, 0, z3.real, z3.imag, **quiver_params)
plt.plot(z1.real, z1.imag, 'ro')
plt.plot(z2.real, z2.imag, 'ro')
plt.plot(z3.real, z3.imag, 'ro')
plt.plot([z1.real, z3.real], [z1.imag, z3.imag], '--')
plt.plot([z2.real, z3.real], [z2.imag, z3.imag], '--')
font_params = dict(fontsize=12, usetex=True)
plt.text(z1.real, z1.imag, r'$z_1$', **font_params, ha='left')
plt.text(z2.real, z2.imag, r'$z_2$', **font_params, ha='right')
plt.text(z3.real, z3.imag, r'$z_1 + z_2$', **font_params, va='bottom')
plt.axis('equal')
plt.savefig('1-1-1.png', bbox_inches='tight', dpi=600)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import cmath

z1, z2 = cmath.exp(complex(0, np.pi/4)), 2*cmath.exp(complex(0, np.pi))
z3 = z1*z2
quiver_params = dict(angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, z1.real, z1.imag, **quiver_params)
plt.quiver(0, 0, z2.real, z2.imag, **quiver_params)
plt.quiver(0, 0, z3.real, z3.imag, **quiver_params, color='C0')
plt.plot(z1.real, z1.imag, 'ro')
plt.plot(z2.real, z2.imag, 'ro')
plt.plot(z3.real, z3.imag, 'ro')
font_params = dict(fontsize=12, usetex=True)
plt.text(z1.real, z1.imag, r'$z_1$', **font_params, ha='left')
plt.text(z2.real, z2.imag, r'$z_2$', **font_params, ha='right')
plt.text(z3.real, z3.imag, r'$z_1\cdot z_2$', **font_params, va='top')
plt.plot([0, 2], [0, 0], 'k--')
theta1, r1 = np.pi/4, 0.6
theta2, r2 = np.pi, 0.4
theta3, r3 = 5*np.pi/4, 0.2
plt.plot(r1*np.cos(np.linspace(0, theta1)), r1*np.sin(np.linspace(0, theta1)), 'k')
plt.plot(r2*np.cos(np.linspace(0, theta2)), r2*np.sin(np.linspace(0, theta2)), 'k')
plt.plot(r3*np.cos(np.linspace(0, theta3)), r3*np.sin(np.linspace(0, theta3)), '--')
plt.text(r1*np.cos(theta1/2), r1*np.sin(theta1/2), r'$\theta_1$', **font_params, va='bottom')
plt.text(r2*np.cos(theta2/2), r2*np.sin(theta2/2), r'$\theta_2$', **font_params, va='bottom')
plt.text(-r3*np.cos(theta3/2), -r3*np.sin(theta3/2), r'$\theta_3$', **font_params, va='top')
plt.xlim([-2.5, 2.5])
plt.ylim([-2.5, 2.5])
plt.gca().set_aspect('equal')
plt.savefig('1-1-2.png', bbox_inches='tight', dpi=600)
```
